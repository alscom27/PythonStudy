```python
import numpy as np

# matplotlib.pyplot          

import matplotlib.pyplot as plt


#   - 'Malgun Gothic'
plt.rcParams["font.family"] = "Malgun Gothic"

#   
plt.rcParams["axes.unicode_minus"] = False

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title("  ")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

labels = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.bar(labels, values)
plt.title("  ")
plt.show()

x = [1, 2, 3, 4, 5]
y = [2, 1, 4, 3, 7]

plt.scatter(x, y)
plt.title(" ")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

labels = ["Apple", "Banana", "Cherry", "Date"]
sizes = [30, 25, 25, 20]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("  ")
plt.axis("equal")  #   
plt.show()
```
