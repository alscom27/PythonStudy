```python
import numpy as np

# matplotlib.pyplot 는 데이터를 시각화 할때 많이 쓰임 주로 그래프나 차트를 그림

import matplotlib.pyplot as plt


# 윈도우 기준 - 'Malgun Gothic'
plt.rcParams["font.family"] = "Malgun Gothic"

# 마이너스 깨짐 방지
plt.rcParams["axes.unicode_minus"] = False

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title("선 그래프 예시")
plt.xlabel("X축")
plt.ylabel("Y축")
plt.grid(True)
plt.show()

labels = ["A", "B", "C", "D"]
values = [10, 20, 15, 25]

plt.bar(labels, values)
plt.title("막대 그래프 예시")
plt.show()

x = [1, 2, 3, 4, 5]
y = [2, 1, 4, 3, 7]

plt.scatter(x, y)
plt.title("산점도 예시")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

labels = ["Apple", "Banana", "Cherry", "Date"]
sizes = [30, 25, 25, 20]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("파이 차트 예시")
plt.axis("equal")  # 원 모양으로 만들기
plt.show()
```
