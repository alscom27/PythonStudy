```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(h, w, maxit=20):
    """Returns an image of the Mandelbrot fractal of size (h, w)."""
    y, x = np.ogrid[-1.4:1.4:h*1j, -2:0.8:w*1j]
    c = x + y * 1j
    z = c
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + c
        diverge = z * np.conj(z) > 2**2  # Who is diverging
        div_now = diverge & (divtime == maxit)  # Who is diverging now
        divtime[div_now] = i  # Note when
        z[diverge] = 2  # Avoid diverging too much

    return divtime

# Mandelbrot fractal 시각화
plt.imshow(mandelbrot(400, 400), cmap="hot")
plt.colorbar()
plt.title("Mandelbrot Fractal")
plt.show()
```

```python
# 1부터 25까지의 숫자로 이루어진 배열을 생성하세요.
# 배열의 각 값을 0과 1 사이로 정규화(normalization)하세요.
# 정규화란 주로 0에서 1사이의 범위로 데이터를 변환하는 작업
# 정규화 공식: (x - min) / (max - min)

import numpy as np

# 1부터 25까지의 배열 생성
array = np.arange(1, 26)

# 최소값과 최대값 계산
min_value = array.min()
max_value = array.max()

# 정규화 수행
normalized_array = (array - min_value) / (max_value - min_value)

print("원본 배열:")
print(array)
print("\n정규화된 배열:")
print(normalized_array)

plt.figure(figsize=(8, 4))
plt.plot(array, normalized_array, marker='o', linestyle='-', color='blue')
plt.title("Normalized Array Plot")
plt.xlabel("Original Value")
plt.ylabel("Normalized Value")
plt.grid(True)
plt.show()
```

```python
# 5 * 5 크기의 배열을 생성하고 배열의 값은 0에서 50 사이의 랜덤 정수로 구성됨.
# 생성한 배열에서 30보다 큰 값만 필터링하여 새 배열을 만드세요
# 필터링된 배열의 값 중 최대값, 최소값, 평균값을 계산하세요.

import numpy as np
import matplotlib.pyplot as plt

# 1. 5x5 크기의 랜덤 배열 생성 (값은 0~50)
matrix = np.random.randint(0, 51, (5, 5))

# 2. 30보다 큰 값 필터링
filtered = matrix[matrix > 30]

# 3. 필터링된 값의 최대값, 최소값, 평균값 계산
max_value = filtered.max() if filtered.size > 0 else None
min_value = filtered.min() if filtered.size > 0 else None
mean_value = filtered.mean() if filtered.size > 0 else None

# 결과 출력
print("5x5 Random Matrix:")
print(matrix)
print("\nFiltered Values Greater Than 30:")
print(filtered)
print("\nMaximum Value of Filtered Values:", max_value)
print("Minimum Value of Filtered Values:", min_value)
print("Mean Value of Filtered Values:", mean_value)

# 4. 시각화 - 원본 배열과 필터링된 값
plt.figure(figsize=(12, 6))

# 원본 배열의 히트맵 시각화
plt.subplot(1, 2, 1)
plt.imshow(matrix, cmap="viridis", interpolation="nearest")
plt.title("Original 5x5 Matrix")
plt.colorbar(label="Value Magnitude")
plt.xlabel("Column")
plt.ylabel("Row")

# 이 히트맵은 원본 5x5 배열의 각 요소의 값을 색상의 강도로 표현합니다.
# 이를 통해 데이터의 분포와 값의 패턴(예: 높은 값과 낮은 값의 위치)을 쉽게 파악할 수 있습니다.

# 필터링된 값을 막대 그래프로 표시
plt.subplot(1, 2, 2)
plt.bar(range(len(filtered)), filtered, color="orange", alpha=0.7)
plt.title("Values Greater Than 30")
plt.xlabel("Index of Filtered Values")
plt.ylabel("Value")

# 이 막대 그래프는 30보다 큰 요소들만을 나열하여 각 값의 크기를 비교할 수 있게 해 줍니다.
# 시각적으로 값들의 차이를 확인하여, 데이터의 범위 및 분포를 한눈에 볼 수 있습니다.

plt.tight_layout()
plt.show()
```

```python
# numpy를 사용해 숫자 1부터 50까지의 배열을 생성하세요.
# 배열에서 짝수만 추출해 새로운 배열을 만드세요.
# 원본 배열과 추출한 배열을 막대 그래프로 시각화하세요.

import numpy as np
import matplotlib.pyplot as plt

# 1. 숫자 1부터 50까지의 배열 생성
array = np.arange(1, 51)

# 2. 짝수만 추출
even_array = array[array % 2 == 0]

# 3. 시각화
plt.figure(figsize=(8, 4))

# 원본 배열 시각화
plt.bar(range(len(array)), array, alpha=0.6, label="Original", color="blue")

# 짝수 배열 시각화
plt.bar(range(len(even_array)), even_array, alpha=0.8, label="Even Array", color="orange")

plt.title("Original Array and Even Array")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()
```
