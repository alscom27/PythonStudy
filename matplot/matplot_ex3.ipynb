```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, np.pi * 2, 100)  # 2*pi는 360도에 해당함
fig = plt.figure()  # 이미지로 저장하기 위해 plt의 figure객체 생성
plt.title('sin cos curve')
plt.plot(x, np.sin(x), "r-", label='sin curve')
plt.plot(x, np.cos(x), "b:", label='cos curve')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend(loc='upper right')
print(plt.style.available) # matplotlib에서 사용할 수 있는 스타일들의 목록
plt.style.use('seaborn-v0_8-whitegrid')
plt.show()

# plot(x,y, 속성(없으면 디폴트인듯..))
# axis(x시작점, x끝점, y시작점, y끝점)

plt.plot([2,3,4], [4,9,16], c='r')
plt.axis([0,20,0,20])
plt.show()
```

```python
import matplotlib.font_manager as fm
from matplotlib import rc # 폰트설정을 위해서 임포트

# 시스템에 설치된 모든 폰트 이름 출력
for font in fm.fontManager.ttflist:
    if 'Nanum' in font.name or 'Gothic' in font.name or 'Malgun' in font.name:
        print(font.name)
        
# 한글폰트 설정
plt.rc('font', family='Malgun Gothic')
# 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

titleFont = {'fontsize': 20, 'color': '#0000ff'}

tri_x_axis = np.linspace(-4,4,100)
sin_y_axis = np.sin(tri_x_axis)
cos_y_axis = np.cos(tri_x_axis)
plt.plot(tri_x_axis, sin_y_axis)
plt.plot(tri_x_axis, cos_y_axis)
plt.title("Trigonometric Functions 삼각함수", fontdict=titleFont)
plt.show()
```

```python
# subplot(행, 열 , 위치)
# 마이너스 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False


plt.subplot(211).set(xticks=[-4,-2,0,2,4], yticks=[-1,0,1])
tri_x_axis = np.linspace(-4,4,100)
sin_y_axis = np.sin(tri_x_axis)
cos_y_axis = np.cos(tri_x_axis)
plt.plot(tri_x_axis, sin_y_axis)
plt.plot(tri_x_axis, cos_y_axis)
plt.title("trigonometric function 삼각함수")

plt.subplot(212).set(xlim=[0,8], ylim=[0,3])
log_x_axis = np.linspace(1,10,100)
log_y_axis = np.log(log_x_axis)
plt.plot(log_x_axis,log_y_axis)
plt.title('log function 로그함수')

plt.show()
```

```python
fig, ax = plt.subplots(2,2) # 2행 2열의 그림을 그리는 공간을 만듬

x = np.random.randn(100) # 정규 분포를 가지는 데이터
y = np.random.randn(100) # ''
ax[0,0].scatter(x,y) # 산점도 그림
x = np.arange(10)
y = np.random.uniform(1,10,10) # 균일분포값 생성
ax[0,1].bar(x,y)
x = np.linspace(0,10,100)
y = np.cos(x)
ax[1,0].plot(x,y) # 실선으로 함수를 그림
z = np.random.uniform(0,1,(5,5))
# imshow는 기본이 gray임 그래서 cmap속성 추가
ax[1,1].imshow(z, cmap='viridis') # 분포를 2D 이미지로 그림

# scatter() : 산점도 그림을 그리는 기능으로 2차원 좌표에 흩뿌려진 점들의 좌표로 자료값을 나타냄.
# bar() : 막대 그래프를 그리는 기능 (수직 막대)
# barh() : 수평 막대 그래프를 그리는 기능
# pie() : 파이 형태의 그림을 그리는 기능
# imshow() : 이미지를 화면에 그리는 기능
# plot() : 실선이나 점선으로 데이터를 나타냄.

# title : axes의 제목
# suptitle : figure의 제목
# figtext : figure 임의의 위치에 텍스트 추가
# text : axes 임의의 위치에 텍스트 추가
# annoatate : axes 임의의 위치에 화살표와 함께 주석 추가
```

```python
n = 30 # 전체 점의 개수
x = np.random.rand(n) # 점들의 임의의 x 좌표
y = np.random.rand(n) # 점들의 임의의 y 좌표
colors = np.random.rand(n) # 랜덤한 색상
area = (30 * np.random.rand(n))**2 # 반지름이 0에서 15포인트 범위

plt.scatter(x,y, s=area, c = colors, alpha=0.5, cmap='viridis')
```

```python
x = np.arange(3)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777,1046,1681]

plt.bar(x, domestic)
plt.bar(x, foreign)
plt.xticks(x, years) # x축으로 연도 눈금값 표시
plt.show()
```

```python
# barh()는 수평막대그래프 함수
# 위의 제주도 관광객 방문 그래프 그림을 barh()함수를 사용하여 그려보자

d_x= np.arange(2,7,2)
f_x = np.arange(1,6,2)
years = ['2010', '2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777,1046,1681]

plt.barh(d_x, domestic)
plt.barh(f_x, foreign, color='orange')
plt.yticks(d_x, years) # x축으로 연도 눈금값 표시
plt.show()

x = np.arange(3)
plt.bar(x, domestic, width=0.25)
plt.bar(x+0.3, foreign, width=0.25, color='orange')
plt.xticks(x,years)
plt.show()

plt.barh(x, domestic, height=0.25)
plt.barh(x-0.3, foreign, height=0.25, color='orange')
plt.yticks(x,years)
plt.show()

```

```python
data = [5,4,6,11]
clist = ['cyan', 'gray', 'orange', 'red']
plt.pie(data, autopct='%.2f', colors=clist)
plt.show()

explode = [.06,.07,.08,.09]
plt.pie(data, autopct='%.2f%%', colors=clist, labels=clist, explode=explode)
plt.show()
```
